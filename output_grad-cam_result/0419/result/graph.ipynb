```python
# defining the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

# No of Data points
N = 950

# initializing random values
# data = np.random.randn(N)

data = pd.read_excel("./900_org_iou_smaller.xlsx")
y=data['IOU']

data2 = pd.read_excel("./900_org_iou_bigger.xlsx")
y2=data2['IOU']

# getting data of the histogram
count, bins_count = np.histogram(y, bins=10)

# finding the PDF of the histogram using count values
pdf = count / sum(count)

# using numpy np.cumsum to calculate the CDF
# We can also find using the PDF values by looping and adding
cdf = np.cumsum(pdf)


count2, bins_count2 = np.histogram(y2, bins=10)
pdf2 = count2 / sum(count2)
cdf2 = np.cumsum(pdf2)


plt.figure(figsize=(8,5))
font = {'weight': 'normal', 'size': 15}
# font2=FontProperties(size=25)
plt.tick_params(labelsize=15)
plt.title("CDF of IoU", fontProperties=FontProperties(size=18)) 
plt.xlabel("IoU Value", fontProperties=FontProperties(size=15)) 
plt.ylabel("Percentile", fontProperties=FontProperties(size=15)) 

plt.grid(ls=':', color='gray', alpha=0.6)

# plotting PDF and CDF
# plt.plot(bins_count[1:], pdf, color="red", label="PDF in Smaller Scenarios")
plt.plot(bins_count[1:], cdf, label="CDF in Smaller Scenarios")

# plt.plot(bins_count2[1:], pdf2, color="red", label="PDF in Smaller Scenarios")
plt.plot(bins_count2[1:], cdf2, label="CDF in Bigger Scenarios")

# plt.legend()
plt.legend(loc=2, prop=font, bbox_to_anchor=(0.55, 0.55)) #handles=[l,l2],labels=['IOU values, IOU mean per 50 x'],


plt.savefig('cdf_IoU.png',bbox_inches = 'tight')
plt.show()
```
